## 🔰 Functions in 60 seconds *(just enough for today)*

A **function** is a recipe with a name. You give it ingredients (**input**), it does
its steps (**body**), and hands you back the dish (**output**).

```
 ingredients ──▶  🍳 recipe steps  ──▶  dish
 arguments   ──▶  def my_func(): ──▶  return value
```

In Python:

```python
def greet(name):                  # 1. DEFINE the recipe (nothing runs yet!)
    return f"Arise, {name}!"      # 2. return = hand back the result

hello = greet("Hunter")           # 3. CALL it — NOW the body runs
```

Only **four things** to remember today:

1. `def name(parameter):` — defines it. The indented lines are the body.
2. Defining does **nothing** by itself. The body runs when you **call** it: `greet("Hunter")`.
3. `return` hands the result back so other code can use it. `print` only *shows* it on screen.
4. Same recipe, different ingredients → different dishes. One `def`, endless reuse.

**Why learn this on Day 1?** Every exercise below is written inside a function.
Use them daily and by Week 6 (when we build gradient descent) functions will feel
like breathing. The *deep* study — scope, `lambda`, `*args` — is Day 3. Not today.

Run the demo and check every line makes sense:

````python
# Demo — run me first (Shift+Enter)

def greet(name):
    """Return a greeting for `name`. (This text documents the recipe.)"""
    return f"Arise, {name}!"

def add(a, b):
    """Return the sum of two numbers."""
    return a + b

print(greet("Hunter"))          # call #1
print(greet("Shadow Monarch"))  # call #2 — SAME recipe, new ingredient
print(add(2, 3))

# ── return vs print: the classic trap ──────────────────────────
result = add(10, 5)   # `result` now HOLDS 15, because add() RETURNS it
print(result * 2)     # so we can keep cooking with it → 30
# If add() had only PRINTED 15, `result` would hold None (nothing),
# and result * 2 would crash. return = usable. print = just visible.
````
