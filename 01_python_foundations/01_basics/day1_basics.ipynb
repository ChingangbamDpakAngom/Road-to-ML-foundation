## 🔰 Functions in 60 seconds *(just enough to use them today)*

A **function** is a named block of code you can run whenever you want, with whatever
input you want.

```python
def greet(name):              # def = "define" · name of function · (parameter)
    message = f"Arise, {name}!"   # the body — indented, runs when CALLED
    return message                # return = hand the result back to the caller

hello = greet("Hunter")       # calling it: "Hunter" is the argument → goes into `name`
print(hello)                  # Arise, Hunter!
```

Four things to remember — that's all you need today:

1. **`def name(parameters):`** defines it; the indented body is what it does.
2. Defining runs **nothing**. The body runs only when you **call** it: `greet("Hunter")`.
3. **`return`** sends a value *back* — `print` only shows something on screen.
   A function without `return` gives back `None`.
4. Same function, different arguments → different results. That's the reuse superpower.

*(Deeper topics — `*args`, scope, lambdas — are Day 3. Ignore them for now.)*

Run the demo below and make sure the output makes sense to you:

````python
# Demo — run me first (Shift+Enter)

def greet(name):
    """Return a greeting for `name`. (This string documents the function.)"""
    return f"Arise, {name}!"

def add(a, b):
    """Return the sum of two numbers."""
    return a + b

print(greet("Hunter"))          # calling with "Hunter"
print(greet("Shadow Monarch"))  # SAME code, different input — reuse!
print(add(2, 3))

# `return` vs `print` — the classic beginner trap:
result = add(10, 5)   # result now HOLDS 15, because add() returns it
print(result * 2)     # we can keep computing with it → 30
# If add() had used print instead of return, `result` would be None
# and result * 2 would crash. Return values let functions feed each other.
````
