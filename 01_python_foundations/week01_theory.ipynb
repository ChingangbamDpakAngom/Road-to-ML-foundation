# 📅 Day 1 — Setup, Variables & Strings

## 📋 Today's Map
| Topic | In plain words | Why it matters in ML later |
|-------|----------------|-----------------------------|
| **Variables** | Named boxes that hold a value | Every dataset & model lives in one |
| **Dynamic typing** | The box learns its type from what you put in | Wrong type = the #1 beginner bug |
| **Core types** | `int`, `float`, `str`, `bool` | Features are numbers; labels are often text |
| **f-strings** | Values inside text: `f"acc: {x:.2%}"` | Printing results humans can read |
| **String methods** | Built-in text tools: `.strip`, `.split`, `.join` | Cleaning messy data starts here |

## Theory

**A variable is a name bound to a value.** Python is *dynamically typed*: you never
declare a type, the value carries it. `x = 5` makes `x` an `int`; `x = "hi"` re-binds
the same name to a `str`. Convenient — and the source of the classic bug below.

**The four core types you meet today**
- `int` — whole numbers (`42`), unbounded in Python.
- `float` — decimals (`3.14`); *approximate* (so `0.1 + 0.2 != 0.3` exactly).
- `str` — text (`"food"`), a sequence of characters.
- `bool` — `True` / `False` (really a special `int`: `True == 1`).

**⚠️ The #1 Day-1 bug:** `input()` **always returns a `str`**. So `"50" == 50` is
`False`, and `"12.5" + 3` explodes. You must convert: `int(raw)` / `float(raw)`.
Check any value's type with `type(x)`.

**f-strings** put values inside text and *format* them:
- `f"{amount:.2f}"` → 2 decimal places (`12.50`)
- `f"{share:.1%}"` → percentage (`0.25` → `25.0%`)
- `f"{name:<10}"` → left-align in a 10-char column (great for tidy tables)

**String methods return NEW strings** (strings are immutable). The ones you'll lean
on all phase: `.strip()`, `.lower()` / `.upper()`, `.split(sep)`, `sep.join(list)`,
`.replace(a, b)`. Slicing works too: `text[::-1]` reads "all of it, backwards."

````python
# 🔎 Read & run — watch types and formatting behave
raw = "12.5"                 # imagine this came from input()
print(type(raw), raw + raw)  # str + str = concatenation, NOT math -> "12.512.5"

amount = float(raw)          # convert text -> number
print(type(amount), amount * 2)          # now it's math -> 25.0

category = "  Food  "
print(f"[{category.strip().lower():<8}] £{amount:.2f}")  # cleaning + formatting
print("2026-07-22".split("-"))           # -> ['2026', '07', '22']
````

# 📅 Day 2 — Control Flow

## 📋 Today's Map
| Topic | In plain words | Why it matters |
|-------|----------------|----------------|
| **`if/elif/else`** | Code that makes decisions | "if loss stops falling → stop training" |
| **Truthiness** | Values that count as True/False on their own | `if data:` guards empty datasets |
| **`for` / `while`** | Repeat over items / repeat until done | Process a million rows, not a million lines |
| **`range`** | A lazy number sequence | First taste of memory-safe iteration |
| **`break` / `continue`** | Stop early / skip this one | Early exit, filtering |

## Theory

**`if / elif / else` — order matters.** Only the *first* true branch runs; the rest
are skipped. So the *most specific* condition must come first. Classic trap: checking
`i % 3` before `i % 15` means 15 matches "Fizz" and never reaches "FizzBuzz".

**Truthiness — what counts as `False` without writing `== ...`:**
`False`, `None`, `0`, `0.0`, `""`, `[]`, `{}`, `()`, `set()`. *Everything else is
truthy.* That's why `if my_list:` means "if the list is non-empty" — idiomatic Python,
better than `if len(my_list) > 0:`.

**`for` vs `while`**
- `for x in iterable:` — when you know *what to loop over* (a list, a range, a string).
- `while condition:` — when you loop *until something changes* (a game, retry logic).
  Guard against infinite loops: something inside must eventually make the condition `False`.

**`range(n)` is lazy** — it does **not** build a list of `n` numbers in memory; it
yields them one at a time. `range(1_000_000_000)` costs almost nothing until you
iterate. `list(range(n))` forces all of them into memory. This lazy idea returns in a
big way with **generators** (Week 3) and streaming datasets (later, to avoid OOM).

**`break`** leaves the loop entirely; **`continue`** skips to the next iteration.
**`enumerate(items)`** gives you `(index, value)` pairs — cleaner than tracking a
counter by hand.

````python
# 🔎 Read & run — truthiness, range laziness, enumerate
print(bool([]), bool([0]), bool(""), bool("0"))   # False True False True  (careful!)

r = range(1_000_000)          # instant — nothing built yet
print(type(r), r[10])         # indexable, but lazy

for i, amount in enumerate([12.5, -3, 0, 40]):
    if amount <= 0:
        continue              # skip non-positive
    print(f"row {i}: £{amount:.2f}")
````

# 📅 Day 3 — Functions & Scope

## 📋 Today's Map
| Topic | In plain words | Why it matters |
|-------|----------------|----------------|
| **`def` / `return`** | A named recipe: input → work → output | ALL real code is built from these |
| **Default args** | Fallback values for parameters | Convenient — and one famous trap |
| **`*args` / `**kwargs`** | Accept *any* number of args | Flexible APIs, wrappers, decorators (Week 3) |
| **LEGB scope** | Where Python looks up a name | Explains "why can't I see that variable?" |

## Theory

**`def name(params): ... return value`.** Defining runs *nothing*; the body runs when
you **call** it. `return` hands a value back so other code can use it — `print` only
*shows* it. A function that only prints "returns" `None`.

**Parameters vs arguments:** parameters are the names in the `def`; arguments are the
actual values you pass at the call site. Functions can `return` a tuple to hand back
several values: `return mean, lo, hi` → unpack with `m, lo, hi = stats(...)`.

**⚠️ The mutable-default-argument trap.** `def add(item, acc=[])` looks innocent but
the *same* list is reused across every call — it accumulates across calls, a real bug.
The default is evaluated **once**, at definition time, not per call. The fix:
```python
def add(item, acc=None):
    if acc is None:
        acc = []
    acc.append(item)
    return acc
```

**`*args` / `**kwargs`.** `*args` collects extra positional arguments into a tuple;
`**kwargs` collects extra keyword arguments into a dict. `def stats(*numbers)` accepts
any count; call it on a list with `stats(*my_list)` (the `*` unpacks). You'll see this
everywhere in library code and again when you write decorators.

**LEGB — the name-lookup order:** **L**ocal → **E**nclosing → **G**lobal → **B**uilt-in.
Python searches those scopes in order. Assigning inside a function makes a *local* name;
to rebind an outer name use `global` (module level) or `nonlocal` (enclosing function) —
both are usually a smell, prefer returning values instead.

````python
# 🔎 Read & run — the mutable-default bug, live
def buggy(item, acc=[]):     # DON'T do this
    acc.append(item)
    return acc

print(buggy("a"))            # ['a']
print(buggy("b"))            # ['a', 'b']  <-- leaked across calls!

def stats(*numbers):         # *args in action
    return sum(numbers) / len(numbers), min(numbers), max(numbers)

print(stats(*[12.5, 40, 3])) # (18.5, 3, 40)
````

# 📅 Day 4 — Collections as Records *(just enough)*

> Full data-structures deep dive is **Week 2**. Today: only what the tracker needs —
> a **list of expense dicts**.

## 📋 Today's Map
| Topic | In plain words | Why it matters |
|-------|----------------|----------------|
| **List** | Ordered, changeable sequence | The "table" of your expenses |
| **Dict** | Key → value lookup | One expense = one dict (a "record") |
| **Record vs table** | dict = row, list-of-dicts = table | Exactly how a DataFrame is shaped later |

## Theory

**Lists** are ordered and mutable: `append`, index, slice, iterate. Big-O worth knowing
even now:
- append / index by position → **O(1)**
- `x in my_list` (membership) / search → **O(n)** (scans every item)

**Dicts** map **keys → values** with **average O(1)** lookup, insert, and membership —
because keys are *hashed* to a location rather than searched one-by-one. That O(1) is
why `x in my_dict` crushes `x in my_list` for lookups, and why the DSA problems this
week (Two Sum, anagrams) all reach for a dict/set. Keys must be *hashable* (immutable:
strings, numbers, tuples — not lists).

**`d["x"]` vs `d.get("x")`:** `d["x"]` raises `KeyError` if missing; `d.get("x")`
returns `None` (or a default you pass: `d.get("x", 0)`). Safer when a key might be absent.

**The modelling idea that carries the whole phase:**
- one **record** = one `dict` → `{"amount": 12.5, "category": "food", "date": "2026-07-22"}`
- a **table** = a `list` of those dicts
This is *literally* the shape a Pandas DataFrame gives you in Phase 3 — you're building
the intuition by hand first.

````python
# 🔎 Read & run — a table is a list of dict records
expenses = [
    {"amount": 12.5, "category": "food",   "date": "2026-07-22"},
    {"amount": 40.0, "category": "travel", "date": "2026-07-22"},
    {"amount":  3.0, "category": "food",   "date": "2026-07-23"},
]

total = sum(e["amount"] for e in expenses)               # O(n) scan
food  = sum(e["amount"] for e in expenses if e["category"] == "food")
cats  = {e["category"] for e in expenses}                # distinct -> a set
print(f"total £{total:.2f} · food £{food:.2f} · categories {cats}")
print(expenses[0].get("note", "no note"))                # .get with a default
````

# 📅 Day 5 — Structuring the Tracker *(design theory)*

## Theory

Today is less *new syntax*, more *how professionals arrange code*. You'll build
`expense_tracker/expenses.py` — the first real deliverable.

**Decompose into small, single-purpose functions.** One function = one job:
`make_expense`, `add_expense`, `list_expenses`, `summary`. Small functions are easier
to **test** (Week 3 = pytest), **reuse**, and **read**. A 200-line `main()` is a smell.

**Separate "talk to the user" from "logic".** `add_expense(expenses, ...)` just changes
data and returns it — no `print`, no `input`. The `main()` menu loop does the talking.
This split is the same one every real program uses (and it's what makes logic testable
without a human typing).

**The menu-loop pattern:**
```
while True:
    choice = input("add / list / summary / quit > ").strip().lower()
    if choice == "quit":
        break
    elif choice == "add":
        ...        # call add_expense
    elif ...
    else:
        print("unknown command")
```

**Scope honesty:** the tracker is **in-memory only** this week — close it and the
expenses vanish. That's deliberate. **JSON persistence is Week 2**, and it's much
easier to bolt on once the in-memory version is clean.

# 📅 Day 6 — Validation & `argparse` CLI

## 📋 Today's Map
| Topic | In plain words | Why it matters |
|-------|----------------|----------------|
| **Boundary validation** | Check input where it enters the program | Bad data never reaches your logic |
| **`try / except`** | Catch an error instead of crashing | A friendly message beats a traceback |
| **`argparse`** | Turn a script into a real CLI tool | The ROADMAP exit target: an `add-expense` command |

## Theory

**Validate at the boundary.** Trust nothing a user types. The moment a value enters
(the `add` command), check it: is the amount a number? positive? is the category
non-empty? Reject bad input with a *helpful* message, so the rest of your code can
assume clean data. This "validate at the edge" habit becomes **Pydantic** in Phase 2.

**Basic `try / except`** (full exception handling is Week 2 — today just the essential):
```python
try:
    amount = float(raw)
except ValueError:
    print(f"'{raw}' isn't a valid amount — try a number like 12.50")
    return
```
Catch the *specific* error (`ValueError`), never a bare `except:` (it swallows real
bugs, even `Ctrl-C`).

**`argparse` — a script becomes a tool.** It parses `sys.argv`, gives you `--help`
for free, and validates argument shape. **Subcommands** let one program expose several
verbs:
```
python expenses.py add 12.50 food --desc "lunch"
python expenses.py list
```
Why bother over hardcoded variables or `input()`? Because a real tool is *scriptable*,
*self-documenting* (`--help`), and *testable*. Recruiters and pipelines run flags, not
prompts. This is the ROADMAP's Week-1 exit evidence: a working `add-expense` command.

````python
# 🔎 Read (don't run in Jupyter — argparse reads terminal args). The shape to build:
import argparse

parser = argparse.ArgumentParser(description="Expense Tracker CLI")
sub = parser.add_subparsers(dest="command")

add = sub.add_parser("add", help="add an expense")
add.add_argument("amount", type=float)        # argparse validates the type for you
add.add_argument("category")
add.add_argument("--desc", default="")

sub.add_parser("list", help="list expenses")
# args = parser.parse_args()   # <- in expenses.py, then dispatch on args.command
````
