```python
# side effect
emoticon = "V.V"  # global variable

def main():
    global emoticon
    emoticon = "XD"  # wanted to change global variable using side effect
    say('Is anyone there?')
    say("Hey it's me!")

def say(phrase):
    print(phrase + " " + emoticon) 

# main()
```
